## Class Vec3

This class defines 3D vectors and associated methods.

````python
from mcpi.vec3 import Vec3
````

The operator __+__ used with lists normally concatenates the two lists.

````python
(1, 2, 3) + (1, 1, 1)
````

The operator __*__ used with a scalar just repeats the list.

````python
(1, 2, 3) * 2
````

In order to do vector addition and scalar multiplication the _Vec3_ class has been defined. 

````python
a = Vec3(1, 2, 3)
b = Vec3(1, 1, 1)
````

````python
a.z
````

Adding two vectors.

````python
a + b
````

Subtracting two vectors.

````python
a - b
````

Multiplying a vector with a scalar.

````python
a * 2
````

````python
a * 0.5
````

Accessing the components x, y, and z.

````python
a.x
````

````python
a.y
````

````python
c = a + b
````

````python
a.x = 10
a
````

````python
clone = a.clone()
````

````python
clone
````

There is an in-place version of the addition and the scalar multiplication: __+=__ and __*=__.

````python
a = Vec3(1, 2, 3)
b = Vec3(1, 1, 1)
a += b
print(a, b)
a *= 2
b *= 10
print(a, b)
````

The length of a vector is 
$\sqrt{x^2 + y^2 + z^2}$

a latex formule is enclosed in $...$ $a=$
a latex index is $ x_2$
a latex exposant is $x^{2+x}$
a latex fraction is preced by $ \frac{a+b}{c+d} $

````python
a = Vec3(1, 2, 3)
a.lengthSqr()
````

There is a bug in the `Vec3` module.
```
def length(self):
        return self.lengthSqr** .5
```
should be like this (with brackets added)
```
def length(self):
        return self.lengthSqr()** .5
```


````python
a.lengthSqr() ** 0.5
````

````python
a.length()
````

````python
-a
````

````python
print(a, b, a+b)
````

A function can be mapped to the 3 vector components.

````python
from math import sin
sin(1)
````

````python
a = Vec3(1, 2, 3)
a._map(sin)
a
````

````python
a = Vec3(1, 2, 3)
a.rotateLeft()
a
````

````python
a.rotateRight()
a
````

When rounding 0.5 and above is rounded to the next hight integer.

````python
c = Vec3(1.1, 2.5, 3.7)
c.iround()
c
````

````python
c = Vec3(1.1, 2.5, 3.7)
c.ifloor()
c
````

Iteration over the 3 vector components is possible.

````python
a = Vec3(1, 2, 3)
for i in a:
    print(i)
````

````python
class Vec3:
    def  __eq__(self, rhs):
        self.x == rhs.x and self.y == rhs.y and self.z == rhs.z
````

````python
a = Vec3(1, 2, 3)
````

````python
a == a
````

````python
Vec3.__ne__
````
